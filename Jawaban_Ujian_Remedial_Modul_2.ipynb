```python
def convert_to_num(cols):
    if cols == 'two':
        return 2
    elif cols == 'three':
        return 3
    elif cols == 'four':
        return 4
    elif cols == 'five':
        return 5
    elif cols == 'six':
        return 6
    elif cols == 'eight':
        return 8
    elif cols == 'twelve':
        return 12
```

```python
auto['Num-of-Doors'] = auto['Num-of-Doors'].apply(convert_to_num)
auto['Cylinders'] = auto['Cylinders'].apply(convert_to_num)
```

Now we will try to fill the NaN value for some of the columns.
For the NaN in Num-of-Doors, Bore, and  Stroke columns fill the NaN value with the mean of their columns with the similar value in the Make and Body-Style columns. *e.g if you have NaN in num-of-doors with hatchback renault car then fill the NaN with the num-of-doors mean of the hatchback renault car. 
 

```python
auto_groupby = auto2.groupby(['Make','Body-Style']).mean()
auto_groupby
```

```python
auto_groupby['Num-of-Doors'].loc['mazda'].loc['sedan']
```

```python
def replace_nan_doors(cols):
    if pd.isna(cols['Num-of-Doors']):
        return auto_groupby['Num-of-Doors'].loc[cols['Make']].loc[cols['Body-Style']]
    else:
        return cols['Num-of-Doors']

def replace_nan_bore(cols):
    if pd.isna(cols['Bore']):
        return auto_groupby['Bore'].loc[cols['Make']].loc[cols['Body-Style']]
    else:
        return cols['Bore']

def replace_nan_stroke(cols):
    if pd.isna(cols['Stroke']):
        return auto_groupby['Stroke'].loc[cols['Make']].loc[cols['Body-Style']]
    else:
        return cols['Stroke']
```

```python
auto2['Num-of-Doors'] = auto2[['Num-of-Doors', 'Make', 'Body-Style']].apply(replace_nan_doors, axis = 1)
auto2['Bore'] = auto2[['Bore', 'Make', 'Body-Style']].apply(replace_nan_bore, axis = 1)
auto2['Stroke'] = auto2[['Stroke', 'Make', 'Body-Style']].apply(replace_nan_stroke, axis = 1)
```

Create a new column called MPG-difference where the value is coming from the difference between Highway-MPG and City-MPG, then create another column called Car-Area from multipication of Length, Width, and Height columns.

```python
def mpg_diff(cols):
    diff = cols['Highway-MPG'] - cols['City-MPG']
    return diff

def car_area(cols):
    A = cols['Length'] * cols['Width'] * cols['Height']
    return A
```

```python
auto3['MPG-difference'] = auto3[['Highway-MPG', 'City-MPG']].apply(mpg_diff, axis = 1)
auto3['Car-Area'] = auto3[['Length', 'Width', 'Height']].apply(car_area, axis = 1)
```

Now, use the IQR method to find the row outlier based on the MPG-Difference columns

```python
Q1_amount = auto3['MPG-difference'].describe()['25%']
Q3_amount = auto3['MPG-difference'].describe()['75%']
iqr = Q3_amount - Q1_amount
```

```python
outlier_index = auto3[(auto3['MPG-difference'] < Q1_amount - (1.5 * iqr)) | (auto3['MPG-difference']> Q3_amount + (1.5 * iqr)) ].index
```

```python
not_outlier_index = auto3[(auto3['MPG-difference'] > Q1_amount - (1.5 * iqr)) & (auto3['MPG-difference']< Q3_amount + (1.5 * iqr)) ].index
```

```python
def try_normal(data):
    result = {'Anderson' : {i:j for i,j in zip(anderson(data)[2], anderson(data)[1])}, 'Shapiro': shapiro(data)[1], 
     'K^2': normaltest(data)[1]}
    result['Anderson']['stat'] = anderson(data)[0]
    return result
```

Dari hasil uji korelasi, baik yang menggunakan metode Pearson maupun Spearman diatas untuk melihat korelasi antara feature 'Price' dengan feature lain yang mempunyai tipe number (int atau float), maka dapat diambil kesimpulan:
- Terdapat beberapa feature yang memiliki korelasi cukup kuat yang memiliki nilai korelasi diatas 0.5 dan dibawah -0.5.
- Feature - Feature tersebut adalah 'Wheelbase', 'Length', 'Width', 'Curb-Weight', 'Cylinders', 'Engine-Size', 'Bore', 'Horsepower', 'City-MPG', 'Highway-MPG', dan 'Car-Area'.
- Feature - feature yang memiliki korelasi positif dengan feature 'Price adalah 'Wheelbase', 'Length', 'Width', 'Curb-Weight', 'Cylinders', 'Engine-Size', 'Bore', 'Horsepower', dan 'Car-Area'.
- Feature - feature yang memiliki korelasi negatif dengan feature 'Price adalah 'City-MPG' dan 'Highway-MPG'
- Feature yang memiliki korelasi positif cukup kuat dengan feature 'Price' adalah 'Curb-Weight'
- Feature yang memiliki korelasi negatif cukup kuat dengan feature 'Price' ada;ah 'Highway-MPG'.
- Korelasi positif menunjukkan kedua feature memiliki hubungan yang searah. Sedangkan korelasi negatif menunjukkan kedua feature memiliki hubungan yang berlawanan

```python
from scipy.stats import chi2_contingency
```

```python
chi2_check = []

for i in auto_clean.drop('Price', axis =1).columns:
    if chi2_contingency(pd.crosstab(auto_clean['Price'], auto_clean[i]))[1] < 0.05:
        chi2_check.append('Reject Null Hypothesis')
    else:
        chi2_check.append('Fail to Reject Null Hypothesis')
```

```python
res = pd.DataFrame(data = [auto_clean.drop('Price', axis =1).columns, chi2_check] ).T 
res.columns = ['Column', 'Hypothesis']
res
```

- Dari hasil uji chi2_contingency, maka feature yang mempengaruhi feature 'Price' dapat dilihat tabel diatas yang menolak H0.
- Feature - feature tersebut merupakan merupakan feature yang bertipe float atau int. 
- Maka untuk meilhat pengaruh feature 'Price' dengan feature - feature tersebut kembali lagi melihat uji korelasi sebelumnya yang menggunakan metode Pearson dan 'Spearman'.

Kesimpulan akhir yang didapat adalah feature - feature yang mempengaruhi feature 'Price' adalah Wheelbase', 'Length', 'Width', 'Curb-Weight', 'Cylinders', 'Engine-Size', 'Bore', 'Horsepower', 'City-MPG', 'Highway-MPG', dan 'Car-Area' dengan feature yang memiliki pengaruh cukup kuat adalah feature 'Curb-Weight' secara positif dan feature 'Highway-MPG' secara negatif.
